```python
import psutil
import time
```

```python
def cpu_task_utilisation():
    total = 0
    for i in range(1, 28966):
        total += i
    return total
```

```python
# Measure CPU capacity and time taken
start_time = time.time()
result = cpu_task_utilisation()
end_time = time.time()
```

```python
# Measure memory utilization
process = psutil.Process()
memory_info = process.memory_info()
```

```python
print(f"Total value: {result}")
print(f"Time taken to execute: {end_time - start_time:.3f} seconds")
print(f"CPU Utilisation: {psutil.cpu_percent(interval=1)}%")
print(f"Total Memory Used: {memory_info.rss / (1024 ** 2):.2f} MB")
```
